```python
# Importar las bibliotecas necesarias
import pandas as pd
import glob

# Definir una función que toma un dataframe y un año como entrada y guarda una columna específica en un archivo CSV
def add_mmlluvia(dataframe, anio):
  # Seleccionar la columna deseada y convertirla en un arreglo
  columna = dataframe["ANUAL"].values

  # Convertir el arreglo en un dataframe y guardar en un archivo CSV
  pd.DataFrame(columna).to_csv("mm_anual"+anio+".csv", index=False)

  # Llamar a la función save_mmAnual para unir todos los archivos CSV en uno solo
  save_mmAnual()

# Definir una función que une todos los archivos CSV en un solo archivo
def save_mmAnual():
  # Obtener la lista de archivos CSV en el directorio actual
  archivos_csv = glob.glob("*.csv")
  # Crear una lista vacía para almacenar los dataframes
  dfs = []

  # Iterar sobre cada archivo CSV
  for archivo in archivos_csv:
      # Leer el archivo CSV y convertirlo en un dataframe
      df = pd.read_csv(archivo)
      # Agregar el dataframe a la lista
      dfs.append(df)

  # Concatenar los dataframes en un solo dataframe
  dataframe_final = pd.concat(dfs, axis=0, ignore_index=True)

  # Guardar el dataframe en un archivo CSV
  dataframe_final.to_csv("archivo_final.csv", index=False)

# Definir la función principal del programa
def main():
  for i in range(18):
    anio = str(2003+(i+1))
    # Cargar el dataframe desde un archivo CSV
    dataframe = pd.read_csv("drive/MyDrive/Talenland/precipitaciones/"+anio+".csv")
    # Llamar a la función add_mmlluvia para guardar la columna deseada en un archivo CSV
    add_mmlluvia(dataframe, anio)

# Llamar a la función principal del programa
main()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import joblib

# cargar los datos de temperatura, lluvia, año, entidad y producción de cosechas
# datos = pd.read_csv('datos_cosechas.csv')

# dividir los datos en un conjunto de entrenamiento y prueba
X = datos[['temperatura', 'lluvia', 'anio', 'entidad']]
y = datos['produccion']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# normalizar los datos
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# crear un modelo de regresión lineal
model = LinearRegression()
model.fit(X_train, y_train)

# hacer predicciones en el conjunto de prueba
y_pred = model.predict(X_test)

# evaluar los índices de error
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Error cuadrático medio:", mse)
print("Coeficiente de determinación:", r2)

# guardar el modelo en un archivo
joblib.dump(model, 'modelo_cosechas.pkl')
```

```python
import numpy as np

limpio3 = limpio2
nan = str(limpio3.DISTRITO[3])



def delete_colums(limpio3,nan):

    # función para reemplazar los valores que contienen la letra 'a'
  def replace_value(value):
      if 'nan' in str(value):
          return '***'
      else:
          return value

  # aplicar la función a la columna 'ciudad'
  limpio3['DISTRITO'] = limpio3['DISTRITO'].apply(lambda x: replace_value(x))

  print(limpio3.head(-1))

  # Variables de cardinales
  cadena1 = "Usuarios"
  cadena2 = "Superficie"
  cadena3 = "Lámina"
  cadena4 = "***"

  # Convertir los valores de la columna 'DISTRITO' a tipo str
  limpio3['DISTRITO'] = limpio3['DISTRITO'].astype(str)

  # Filtrar las filas que NO contienen la cadena1 en la columna 'DISTRITO'
  limpio3 = limpio3[~limpio3['DISTRITO'].str.contains(cadena1)]
  limpio3 = limpio3[~limpio3['DISTRITO'].str.contains(cadena2)]
  limpio3 = limpio3[~limpio3['DISTRITO'].str.contains(cadena3)]
  limpio3 = limpio3[~limpio3['DISTRITO'].str.contains(cadena4)]



  
  

delete_colums(limpio3,nan)

# print(limpio3.head(10))
```
